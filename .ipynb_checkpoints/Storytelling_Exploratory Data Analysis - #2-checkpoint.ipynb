## 1. Data Source 

For this project, I used publicly available Electronic Health Records (EHRs) datasets. The MIT Media Lab for Computational Physiology has developed MIMIC-IIIv1.4 dataset based on 46,520 patients who stayed in critical care units of the Beth Israel Deaconess Medical Center of Boston between 2001 and 2012. MIMIC-IIIv1.4 dataset is freely available to researchers across the world. A formal request should be made directly to www.mimic.physionet.org, to gain acess to the data. There is a required course on human research ‘Data or Specimens Only Research’ prior to data acess request. I have secured one here -www.citiprogram.org/verify/?kb6607b78-5821-4de5-8cad-daf929f7fbbf-33486907

The dataset has 26 relational tables including patient’s hospital admission, callout information when patient was ready for discharge, caregiver information, electronic charted events including vital signs and any additional information relevant to patient care, patient demographic data, list of services the patient was admitted or transferred under,  ICU stay types, diagnoses types, laboratory measurments, microbiology tests and sensitivity, prescription data and billing information. 

Although I have full access to the MIMIC-IIIv1.4 datasets, I can not share any part of the data publicly. If you are interested to learn more about the data, there is a MIMIC III Demo dataset based on 100 patients https://mimic.physionet.org/gettingstarted/demo/. If you are interested to requesting access to the data - https://mimic.physionet.org/gettingstarted/access/.

### Results #1: 

+ The final 36 columns(variables) were selected and created from a total of 324 columns found in 26 relational tabeles of MIMIC-IIIv1.4 database.
+ Out of the 5000 randomly selected patients (10% of the total patient population), we have 4,602 unique patients in the final clinical deterioration dataset. Around 400 were lost during data cleaning. MIMIC-IIIv1.4 database has a total population of 46,520 patients.
+ There are 5,849 unique Intensive Care Unit(ICU) stays, some patients were addmitted to ICU more than once. The entire MIMIC-IIIv1.4 datasets has 61,532 ICU stays. Our random sample represent 10% of the total number of ICU stays. That is, ICU stays in this data only includes for the 10% randmly selected patients. The law of large numbers did not disapoint here.
+ There are 800,097 electronic chart events from 4,602 patients. The primary repository of a patients's information is their electronic chart. Each raw in this the final 'Clinical Deterioration' dataset represent a unique entry of patient electronic chart. 
+ The data spans June 2001 - October 2011. In this dataset, the years were moved to future years to comply with HIPAA.
+ 56% of the patients were Male and 44% were female. 
+ The majority of patients (75.4%) were adults (above 18 years old) and 24.6% were children (18 and below). 
+ The majority of patients (69.7%) were whites, African Americans (8.4%), Asians (3.8%), and Hispanic/Lations were (3.4%). Aroundd 12% of the patients did not specified or declined to answer or unable to obtain. 
+ The major type of ICU admission type for adults is due to emergency (61.4%) and for children its newborn (16.7%) admission type. 
+ Most patietns have Medicare (42.2%) and Private (41.5%) insurance. 1.5% of patients self-paid.  
+ The top three services for adults were MED -general service for internal medicine, CMED -for non-surgical cardiac related admiissions, and CSURG - for surgical cardiac admissions and for children the vast majority of patients were neonatal patients. 

## 2. Demonstrate Correlations

```python
# Compute correlations
corr = cd.corr()

# Exclude duplicate correlations by masking uper right values
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set background color / chart style
sns.set_style(style = 'white')

# Set up  matplotlib figure
f, ax = plt.subplots(figsize=(18, 18))

# Add diverging colormap
cmap = sns.diverging_palette(10, 250, as_cmap=True)

# Draw correlation plot
sns.heatmap(corr, mask=mask, cmap=cmap, 
        square=True,
        linewidths=.5, cbar_kws={"shrink": .5}, ax=ax)
ax.set (title='Correlations')
```

```python
# Compute correlations for children only
df=cd[cd['CHILD']==1]
corr = df.corr()

# Exclude duplicate correlations by masking uper right values
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set background color / chart style
sns.set_style(style = 'white')

# Set up  matplotlib figure
f, ax = plt.subplots(figsize=(18, 18))

# Add diverging colormap
cmap = sns.diverging_palette(10, 250, as_cmap=True)

# Draw correlation plot
sns.heatmap(corr, mask=mask, cmap=cmap, 
        square=True,
        linewidths=.5, cbar_kws={"shrink": .5}, ax=ax)
ax.set (title='Correlations: Child Patients')
```

### Results #2: 

The heat map above provides a wholestic picture of where the correlation between variables in the final dataset stands. Bluish-to-deep blue color represents positive correlation between two corrosponding variables. Redish-to-deep red color represents negative correlation between two corrosponding variables (columns). The darker the color, the stronger the correlation. There are large numbers of variables, I will only list few worth mentions here.
+ There is positive correlation between ICU Death (death while admitted at ICU) and WBC (White blood cell counts), respiratory rates, potassium levels, level of urea nitorgen in the blood (BUN), Anion gap (levels of acid in the blood) and age (obviously). Here, positive correlation indicates those who died at ICU charted higher levels during their time at ICU. 
+ There is negative correlation between ICU Death (death while admitted at ICU) and Platelet Count, O2 saturation pulseoxymety, Hemoglobin (amount of hemoglobin the blood), heart rate, Arterial O2 Saturation, and weight during admission. Negative correlation indcates those who died at ICU charted lower leves during their time at ICU.

+ When I run the correlation to only child patients (18 and below), almost all of the positive correlations disappears. The negative correlations remaind the same. 

+ The correlation for adults showed the same results with the whole dataset. In addition, negative correlation with Arterial Blood Pressure systolic emerged.

+ There is negative relationship between age and blood pressure (Diastolic) and positive relationship between age and systolic blood pressure.

+ Patients with longer days of stay at ICU charted higher heart rates (strong positive relationship). 
+ When disagregated by 'CHILD'- (looking at LOS relationship with children and adults separately), the relationship with child patient got more stronger and with adults, it got weaker. 
+ Platelet count and Arterial O2 saturation has negative relationship. 
+ There is a stong negative relationship between length of ICU stay and White blood cell counts among male child patients who died while at ICU.
